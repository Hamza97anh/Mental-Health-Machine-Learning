```python
# Make the "Have you ever been diagnosed with a mental health disorder?" Column have 0 and 1 value. Anything NaN will be treated as zero
def fix_disorder_column():
    merged_data["Have you ever been diagnosed with a mental health disorder?"] = merged_data["Have you ever been diagnosed with a mental health disorder?"].str.lower()
    merged_data["Have you ever been diagnosed with a mental health disorder?"].replace(['No','no'], '0', inplace=True)
    merged_data["Have you ever been diagnosed with a mental health disorder?"].replace(["Yes", "yes"], "1", inplace=True)
    merged_data["Have you ever been diagnosed with a mental health disorder?"].fillna(value="0", inplace=True)

    
fix_disorder_column()
#Should have three categories
merged_data["Have you ever been diagnosed with a mental health disorder?"].unique()
```
